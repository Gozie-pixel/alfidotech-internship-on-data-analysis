```python
CONVERTING PODATE AND RECEIVING DATE TO DATETIME
```

```python
data5['PODate'] = pd.to_datetime(data5['PODate'], errors='coerce')
data5['PODate']
```

```python
data5['ReceivingDate'] = pd.to_datetime(data5['ReceivingDate'], errors='coerce')
data5['ReceivingDate']
```

```python
data5['SupplyDuration'] = (data5['ReceivingDate'] - data5['PODate']).dt.days
Supply_Duration = data5['SupplyDuration']
Supply_Duration.head()
print("Supply Duration (in days):\n",Supply_Duration.head())
```

```python
average_supply_duration = data5['SupplyDuration'].mean()
print("Average Supply Duration (in days):\n",average_supply_duration )
```

```python
data5['InvoiceDate'] = pd.to_datetime(data5['InvoiceDate'])
data5['PayDate'] = pd.to_datetime(data5['PayDate'])

data5['PaymentDuration'] = (data5['PayDate'] - data5['InvoiceDate']).dt.days

average_payment_duration = data5['PaymentDuration'].mean()

print("Average Payment Duration (in days):", average_payment_duration)
```
