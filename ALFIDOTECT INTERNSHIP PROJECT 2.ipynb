# DATA VISUALIZATION

```python
import seaborn as sns
dir(sns)
```

```python
plt.figure(figsize=(15, 6))
sns.lineplot(x='month_name', y='MILES', data=data.groupby('month_name')['MILES'].mean().reset_index())
plt.title('Average miles per Months')
plt.xlabel('Month')
plt.ylabel('miles')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.lineplot(x='day_name', y='MILES', data=data.groupby('day_name')['MILES'].mean().reset_index())
plt.title('Average miles per day')
plt.xlabel('days')
plt.ylabel('miles')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='PURPOSE', data=data, palette='viridis')
plt.xlabel('PURPOSE')
plt.ylabel('Count')
plt.title('purpose Counts')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.histplot(data['PURPOSE'], bins=10, color='green', kde=True)
plt.xlabel('start to stop')
plt.ylabel('frquency')
plt.title('star to stop frequency')
plt.tight_layout()
plt.show()
```
